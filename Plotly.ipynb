```python
import plotly.express as px
```

```python
nz = px.data.gapminder().query("country == 'New Zealand'")
nz.head()
```

```python
fig = px.bar(nz,x='year',y='pop')
fig.show()
```

```python
print(fig)
```

```python
fig.update_layout(template='plotly_dark')
fig.show()
```

```python
iris = px.data.iris()
fig = px.scatter(iris,x='sepal_width',y='sepal_length',color='species')
```

```python
fig.show()
```

```python
fig.write_html('iris.html')
```

```python
nz = px.data.gapminder().query("continent == 'Oceania'")
nz.head()
```

```python
fig = px.line(nz, x='year',y='lifeExp',color='country')
fig.show()
```

```python
fig.update_layout(
    legend = dict(
        yanchor = 'top',
        y=0.95,
        xanchor='left',
        x=0.01),
        title='Average life expectancy'
)
```

```python
fig = px.bar(nz,x='year',y='pop',color='country',barmode='group')
fig.show()
```

```python
fig.update_layout(legend=dict(orientation='h'))
```

```python
df = px.data.gapminder().query("year == '2007'")
df.head()
```

```python
fig = px.scatter(df,x='lifeExp',y='gdpPercap',color='country')
fig.show()
```

```python
fig = px.scatter(df,x='lifeExp',y='gdpPercap',color='country',size='pop')
fig.show()
```

```python
fig = px.scatter(df,x='lifeExp',y='gdpPercap',color='country',size='pop',labels={'lifeExp':'Life Expectation','gdpPercap':'gdpPercapita'})
fig.show()
```

```python
fig.update_traces(marker=dict(line=dict(width=2,color='DarkSlateGrey')))
```
